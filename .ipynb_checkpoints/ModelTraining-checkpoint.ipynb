1) Acceleration and gyro xyz are not oriented in a specific way to the vehicle direction, different booking ID have different orientation.<br>
2) Device could be handled or moved during the journey and affect the device x-y-z accelerations and gyro measurements.<br>
3) Different devices measure differently, the calibration varies.<br>
4) To my knowledge we cannot reorient the xyz axises without the help of magnetometer readings and rotation matrix<br>
5) Data have random missing rows. for examples bookingID: 395136991269 the journey is 1469seconds, but there's only 1080rows.<br>
6) Not all data are reliable. Note 541165879428 where speed are constantly -1, This has been addressed in data cleaning.<br>
7) Bearing become zero when vehicle is stationary. We will need to perform some trick to compute for rate of change of bearing according to time,between moving and stopping.

## FEATURE ENGINEERING
### Thoughts
1)Stated in a few literatures, safe acceleration and decceleration is roughly less than 2m/s. Anything in excess is unsafe driving. <br>
2) Since we have limited data to accurately reorient device x-y-z acceleration and gyro according to vehicle direction, nothing much can be done with lateral acceleration and gyro readings. <br>
3) We can however, take the combined magnitude of x-y-z acceleration, and compute for jerk which is rate of change of acceleration.<br>
4) Rate of change in Bearing aka angular velocity can indicate driver aggressiveness in turning. <br>
5) When all of the above are aggregated for max, mean, and std. Perhaps something meaningful can come out of it. <br>
6) We can multiply Speed to the features, so speed acts as a weight. for example, it is more dangerous to change direction at a faster speed. 

```python
uniqueBookingID=df_all.bookingID.unique()
```

```python
#Set bookingID and second as index and sort them so seconds are in ascending order per booking ID
#This will help us avoid using groupby and sort_values which are time consuming, especially when we do it over loops. 
df_all.set_index(['bookingID','second'],inplace=True)
df_all.sort_index(inplace=True)
#Remove second from index, so we will call the data based on bookingID only
df_all.reset_index(level=1, inplace=True, col_level=3 )
```

**Below code needs 30 minutes to compute on my local machine**

```python
#we will need to use sine and radian function from math module
import math


#Initialize a few objects
df_agg=pd.DataFrame()
mydict={}



#Create features by looping through every bookingID. 
#We need to take difference between subsequent rows within the under the same bookingID for most features, then aggregate them. 
for i,ID in enumerate(uniqueBookingID):

    #Progress bar to track progress
    if i%50 ==0:
        print('progress:',i,'/',len(uniqueBookingID))
    
    df_features=df_all.loc[ID]
    df_features['Time_Interval']= df_features['second']-df_features['second'].shift(1)
      
    
    
    #Feature 1: Vehicle acceleration
    #===============================
    #Computed by checking for Speed now > Speed before, take the absolute value of Speed difference over time interval
    df_features['Vehicle_Accel']= abs((df_features['Speed']>df_features['Speed'].shift(1))*(
        (df_features['Speed']-df_features['Speed'].shift(1))/df_features['Time_Interval']))
    
    #Feature 2: Vehicle decceleration
    #=================================    
    #Computed by checking for Speed now < Speed before, take the absolute value of Speed difference over time interval
    df_features['Vehicle_Deccel']= abs((df_features['Speed']<df_features['Speed'].shift(1))*(
        (df_features['Speed']-df_features['Speed'].shift(1))/df_features['Time_Interval']))
    
    
    #Feature 3: Device Acceleration Magnitude
    #=================================    
    #Using Pythagorean triple, compute for device acceleration magnitude
    df_features['Device_Accel'] = np.sqrt(df_features['acceleration_x']**2+df_features['acceleration_y']**2+
                                          df_features['acceleration_z']**2)
    
    #Feature 4: Device Jerk Magnitude
    #=================================    
    #This is the rate of change of device acceleration magnitude
    df_features['Device_Jerk'] = abs(df_features['Device_Accel']-df_features['Device_Accel'].shift(1))
    
    
    #Feature 5: Rate of Bearing
    #=================================    
    #Bearing is 0 when car is stationary,we will remove stationary vehicle data since it give a erroneous computation.
    #We need to handle the data differently, so a separate dataframe is used in this computation
    df_bearing = df_features[['Speed','Bearing','Time_Interval']].copy()
    df_bearing=df_bearing[df_bearing['Speed']>0]
    df_bearing['Rate_Bearing']=abs(((df_bearing['Bearing'].apply(lambda x: math.sin(math.radians(x)))-df_bearing['Bearing'].
                                     shift(1).apply(lambda x: math.sin(math.radians(x))))/df_bearing['Time_Interval']))
    #Drop anything above (sine(90deg) = 1), as it is unrealistic.
    df_bearing=df_bearing[df_bearing['Rate_Bearing']<1]

    
    #Feature 6-10: Speed Multipliers
    #================================= 
    df_features['Speed-Vehicle_Accel']=abs(df_features['Speed']*df_features['Vehicle_Accel'])
    df_features['Speed-Vehicle_Deccel']=abs(df_features['Speed']*df_features['Vehicle_Deccel'])
    df_features['Speed-Device_Accel']=abs(df_features['Speed']*df_features['Device_Accel'])
    df_features['Speed-Device_Jerk']=abs(df_features['Speed']*df_features['Device_Jerk'])
    df_bearing['Speed-Rate_Bearing']=abs(df_bearing['Speed']*df_bearing['Rate_Bearing'])

    
    
    #Feature 11-13: 2D-Plane Acceleration Magnitude
    #================================= 
    df_features['xy_Device_Accel']=np.sqrt((df_features['acceleration_x']**2)*(df_features['acceleration_y']**2))
    df_features['xz_Device_Accel']=np.sqrt((df_features['acceleration_x']**2)*(df_features['acceleration_z']**2))
    df_features['yz_Device_Accel']=np.sqrt((df_features['acceleration_y']**2)*(df_features['acceleration_z']**2))
    
    
    #Convert Gyro to Absolute value
    #================================= 
    df_features['gyro_x']=abs(df_features['gyro_x'])
    df_features['gyro_y']=abs(df_features['gyro_y'])
    df_features['gyro_z']=abs(df_features['gyro_z'])    
    
    
    
    #Replace NaN with 0. NaN is a result of the shift function. Any NaN value can be safely replaced by 0.
    df_features.fillna(0)
    
    #Save bookingID to dict.
    mydict['bookingID']=ID
    
    #Time_Interval-mode column is for some extra EDA
    mydict['Time_Interval-mode']=df_features['Time_Interval'].value_counts().index[0]
    
    
    #Aggregate all above features into mean,max and std. 
    for function in ['mean','max','std']:
        for item in ['Vehicle_Accel','Speed-Vehicle_Accel']:
            #Set threshold to filter out normal cruising acceleration and deceleration for aggregation
            df_func=df_features[df_features['Vehicle_Accel']>0.5][item].agg(function)
            new_col=item+'-'+function
            mydict[new_col]=df_func
    
        for item in ['Vehicle_Deccel','Speed-Vehicle_Deccel']:
            #Set threshold to filter out normal cruising acceleration and deceleration for aggregation
            df_func=df_features[df_features['Vehicle_Deccel']>0.5][item].agg(function)
            new_col=item+'-'+function
            mydict[new_col]=df_func

        for item in ['Device_Accel','Speed-Device_Accel','Device_Jerk','Speed-Device_Jerk']:
            #Set threshold to filter out normal cruising acceleration and deceleration for aggregation
            df_func=df_features[df_features['Speed']>0][item].agg(function)
            new_col=item+'-'+function
            mydict[new_col]=df_func
        
        for item in ['xy_Device_Accel','xz_Device_Accel','yz_Device_Accel','gyro_x','gyro_y','gyro_z']:
            #Set threshold to filter out normal cruising acceleration and deceleration for aggregation
            df_func=df_features[item].agg(function)
            new_col=item+'-'+function
            mydict[new_col]=df_func
        
        for item in ['Rate_Bearing','Speed-Rate_Bearing']:
            df_func=df_bearing[item].agg(function)
            new_col=item+'-'+function
            mydict[new_col]=df_func
    
    
    df_agg=df_agg.append(mydict,ignore_index=True)    
    

print('progress:',i+1,'/',len(uniqueBookingID),'DONE')
```

We will also setup a function for plotting : <br>
1) Receiver Operating Characteristic (ROC)<br>
2) Precision-Recall curve (PRC)

```python
def plot_ROC_PRC(false_positive_rate, true_positive_rate,recall, precision):
        # Plot ROC curves
        fig= plt.figure(figsize=(15,8))
        ax1=fig.add_subplot(1,2,1)    
        ax1.set_title('Receiver Operating Characteristic')
        ax1.plot(false_positive_rate, true_positive_rate)
        ax1.plot([0, 1], ls="--")
        ax1.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
        ax1.set_ylabel('True Positive Rate')
        ax1.set_xlabel('False Positive Rate')
        
        # Plot PRC curves
        ax2=fig.add_subplot(1,2,2)    
        ax2.set_title('Precision-Recall curve')
        ax2.plot(recall, precision)
        ax2.plot([0, 0], [1,0] , c=".7"), plt.plot([0,0] , c=".7")
        ax2.set_ylabel('Precision')
        ax2.set_xlabel('Recall')
        plt.show()
```

#### 4. Keras Sequential Binary Classification

Setup Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

tf.keras.backend.clear_session()
# create model
sq_model = tf.keras.Sequential()
sq_model.add(Dense(16, input_dim=len(X_train.columns), kernel_initializer='normal', activation='relu'))
sq_model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
# Compile model
sq_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
sq_model.summary()
```

### Cross Validation using StratifiedKFold

Using good judgement, the n_estimators aka iterations is set at 3000 to get the best AUC-ROC score without compromising too much computational speed. 

```python
#Setup Model for 3000 iterations
lgb_model = lgb.LGBMClassifier(objective='binary',metric='auc',boosting_type='goss',class_weight={0:1,1:2.6},
                               learning_rate=0.0012,n_estimators=5000,importance_type='gain',
                              random_state=123,n_jobs=4, min_split_gain=3,num_leaves=13,reg_alpha=0.6,reg_lambda=0.84)
lgb_model
```

We will use 10 Kfold and compute for the mean average precision and AUC-ROC score. 

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state = 123)
score_dict_Kfold={}
kfold_performance=pd.DataFrame()
n=1
for train_index, test_index in skf.split(Xtop10, ytop10):
    # create model
    print('\nKfold Split:',n)
    X_train, X_test = Xtop10.iloc[train_index], Xtop10.iloc[test_index]
    y_train, y_test = ytop10.iloc[train_index], ytop10.iloc[test_index]
       
    start = time.time()
    lgb_model.fit(X_train,y_train, verbose=1000)
    y_score_lgb=lgb_model.predict_proba(X_test)[:,1] 
    end = time.time()
    runtime_lgb = round(end - start,2)
    
    false_positive_rate_lgb, true_positive_rate_lgb, threshold = roc_curve(y_test, y_score_lgb)
    precision_lgb, recall_lgb, _ = precision_recall_curve(y_test, y_score_lgb)

    auc_lgb=roc_auc_score(y_test, y_score_lgb)
    avg_precision_lgb=average_precision_score(y_test, y_score_lgb)
    print('roc_auc_score: ', auc_lgb, '  Average Precision: ',avg_precision_lgb, '  Computation Time: ', runtime_lgb ,'s')
    
    n=n+1
    
    score_dict_Kfold=({'AUC-ROC':auc_lgb, 'Avg-Precision':avg_precision_lgb, 'Computation Time':runtime_lgb})
    kfold_performance=kfold_performance.append(score_dict_Kfold,ignore_index=True)
```

**Tabulated score can be found below:**

```python
print('Light Gradient Boosting Model:')
print('individual score from each kfolds splits')
display(kfold_performance)

print('average score from all kfolds splits')
display(pd.DataFrame(kfold_performance.agg('mean'), columns=['mean']).join(pd.DataFrame(kfold_performance.agg('std'), columns=['std'])))
```
